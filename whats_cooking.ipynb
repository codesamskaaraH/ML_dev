```python
#get ingred count 
def get_ingred_count(df):
    ingred_count = {}
    for i in range(len(df)):
        for j in df['ingredients'][i]:
            if j in ingred_count.keys():
                ingred_count[j]+=1
            else:
                ingred_count[j]=1
    return ingred_count
```

```python
def get_missing_ingred(test_cnt, train_cnt):
    ingred_miss=[]
    for i in test_cnt.keys():
        if i not in train_cnt.keys():
            ingred_miss.append(i)
    return ingred_miss
```

```python
def readjust_df(data, df):
    for i in data:
        df[i]=0
    return df
```

```python
def pr_red(dat):
    print('\033[91m' + str(dat))
```

```python
def prep_ds(ds, data):
    for i in data.keys():
        ds[i]=np.zeros(len(ds))
    return ds
```

```python
def set_ingred(ds):
    for i in range(len(ds)):
        for j in ds['ingredients'][i]:
            ds[j].iloc[i]=1
```
